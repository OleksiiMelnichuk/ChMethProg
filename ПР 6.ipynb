```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.interpolate import lagrange
```

```python
x_i=np.array([-1.,0.,1.,3.]) 
fx_i=np.array([1.,-8.,-3.,25.]) 
point1=-0.5
point2=1.5
point3=2.
point4=2.5

def lagrange_interpol(x_i, fx_i, point): 
    n=len(x_i)
    p=np.zeros(n)
    for i in range(n): 
        p_i=1 
        for j in range(n): 
            if i != j: 
                p_i *= (point-x_i[j])/(x_i[i]-x_i[j]) 
        p[i]=p_i
    return np.dot(fx_i, p)
f_interp=lagrange_interpol(x_i, fx_i, point1) 
print("Значення функції у точці point1 =", f_interp.round(4))
f_interp=lagrange_interpol(x_i, fx_i, point2) 
print("Значення функції у точці point2 =", f_interp.round(4)) 
f_interp=lagrange_interpol(x_i, fx_i, point3) 
print("Значення функції у точці point3 =", f_interp.round(4)) 
f_interp=lagrange_interpol(x_i, fx_i, point4) 
print("Значення функції у точці point4 =", f_interp.round(4)) 
```

```python
x_new=np.linspace(np.min(x_i),np.max(x_i),100)
y_new=[lagrange_interpol(x_i ,fx_i,i) for i in x_new] 
plt.plot(x_i, fx_i, 'o', x_new, y_new)
plt.title('Графік Лагранжа', color='green') 
plt.grid() 
plt.show()  
```

```python
f=lagrange(x_i, fx_i)
fig = plt.figure(figsize = (7,5)) 
plt.plot(x_new, f(x_new), 'b', x_i, fx_i, 'ro') 
plt.title('Перевірка графіка Лагранжа', color='r') 
plt.grid() 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show() 
```
