```python
import numpy as np
from scipy.optimize import least_squares
import matplotlib.pyplot as plt
```

```python
def fun(a, x, y):
    return np.polyval(a, x)-y

def f(x):
    return x**2*np.sin(x)

x=np.arange(0.1, 1.1, 0.1)
y =np.array([f(i) for i in x])
print(x)
print(y)
a0 = np.array([1, 1])
res_lsq = least_squares(fun, x0=a0, args=(x, y))
print("a0 = %.2f, a1 = %.2f" % tuple(res_lsq.x))

x_p = np.linspace(min(x), max(x), 100)
y_p = np.polyval(res_lsq.x, x_p)

plt.plot(x, y, 'o', color="red")
plt.plot(x_p, y_p, color='green')
plt.title("МНК наближення прямою")
plt.grid()
plt.show()
```

```python
x=np.arange(0.1, 1.1, 0.1)
y=f(x)
print(x)
print(y)

def fun(a, x, y):
    return a[0]+a[1]*x+a[2] * x**2 - y

a0 = np.array([1, 1, 1])
res_lsq = least_squares(fun, x0=a0, args=(x, y))
print("a0 = %.2f, a1 = %.2f, a2 = %.2f" % tuple(res_lsq.x))

f=lambda x: sum([u*v for u, v in zip(res_lsq.x, [1, x, x**2])])
x_p=np.linspace(min(x), max(x), 20)
y_p=f(x_p)

plt.plot(x, y, 'o', color="red")
plt.plot(x_p, y_p, color='green')
plt.title("МНК наближення параболою")
plt.grid(True)
plt.show()
```
