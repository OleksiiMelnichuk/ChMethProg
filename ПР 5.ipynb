```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import optimize
from scipy.misc import derivative
```

```python
x_min=-0.3
x_max=-0.1
y_min=0.5
y_max=0.6
step=0.01
x, y=np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
equat_1=np.sin(y+1)-x-1.2
equat_2=2*y+np.cos(x)-2

fig, ax= plt.subplots(figsize=(10, 10))
ax.contour(x, y, equat_1, levels=[0], colors='red')
ax.contour(x, y, equat_2, levels=[0], colors='blue')
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь')
plt.grid()
plt.show()
```

```python
x_0=-0.2
y_0=0.55
eps=0.001

def f1(y):
    return np.sin(y+1)-1.2

def f2(x):
    return 1-0.5*np.cos(x)

def iteration(x, y, eps):
    x_n=x
    y_n=y
    x_n1=f2(x)
    y_n1=f1(y)
    n=1
    while ((abs(x_n1-x_n)>=eps)&(abs(y_n1-y_n)>=eps)):
        x_n=x_n1
        y_n=y_n1
        x_n1=f2(y_n)
        y_n1=f1(x_n)
        n+=1
    print('Проста ітерація')
    print('x=', y_n,'\ny=', x_n, '\nКількість ітерацій:', n)
iteration(x_0, y_0, eps)
```
